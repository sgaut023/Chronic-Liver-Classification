```python
from fastai.vision.all import *
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_pickle('./bmodes_steatosis_assessment_IJCARS.pickle')
```

```python
ttt = [np.arange(1,11).astype('str')] *55
rr = [t for tt in ttt for t in tt]
pid_list = rr
dataset['fname'] = [f"./images/shanel/P{d_id}_image{d_pid}.jpg" for d_id,d_pid in list(zip(dataset['id'],pid_list))]
dataset.rename(columns={'class':'labels'},inplace=True)
```

```python
def get_train_test_patients_id(ids, train_sz: float=.9, seed:int=2020):
    # Get the list of patient id's and shuffle that list
    patient_ids = ids.unique()
    np.random.seed(seed)
    np.random.shuffle(patient_ids)
    # Create two lists of patient id's for training and testing
    train_patient_cnt = int(len(patient_ids) * train_sz)
    train_id = patient_ids[0:int(train_patient_cnt)]
    test_id = patient_ids[int(train_patient_cnt): int(len(patient_ids))]
    return train_id, test_id


def train_test_split(data: pd.DataFrame, train_sz:float=.9, seed:int=2020):
    """
    Function that will split the dataset into training and testing based
    on patient id's. A patient will not appear in both training and testing set.
   
    """
    assert 'id' in data, f'Column id not present in dataframe.'
   
    train_id, test_id = get_train_test_patients_id(data['id'],train_sz , seed)
    # Separate the features from id and class columns
    train_data = data[data['id'].isin(train_id)].reset_index(drop=True)
    test_data = data[data['id'].isin(test_id)].reset_index(drop=True)
    #TO DO: IS IT SHUFFLE?
    return train_data, test_data
```

```python
train_data , test_data = train_test_split(dataset[['fname','labels','id']])
train_data, valid_data = train_test_split(train_data)
```

```python
train_data['is_valid'] = False
valid_data['is_valid'] = True

training_data = pd.concat([train_data,valid_data],axis=0).reset_index(drop=True).sample(frac=1,random_state=2020)
_ = training_data.pop('id')
```

```python
dls = ImageDataLoaders.from_df(training_data, './', folder='./', valid_col='is_valid',bs=16,item_tfms=Resize(224))
```

```python
dls.show_batch()
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(1,lr=10e-3)
```

```python
learn.fine_tune(5,lr=3e-4)
```

```python
preds = [learn.predict(test_data['fname'].iloc[i]) for i in range(test_data.shape[0])]
```

```python
real_preds = [int(pred[0]) for pred in preds]
```

```python
accuracy_score(test_data['labels'].values, np.array(real_preds))
```
